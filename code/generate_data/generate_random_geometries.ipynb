```python
import random
import numpy as np
```

```python
def to_meters(size):
  return size * 10**(-3)
```

```python
def random_vessel_params(type, length, height):

  match type:

    case 'healthy':
      vessel_params = {
        'healthy': {
          'undulation_amplitude': to_meters(np.random.uniform(height/150, height/25)),
          'undulation_frequency': np.random.uniform(1, 3),
          'oscilations': int(np.random.uniform(2, 6)),
        }}

    case 'isthmus':
      vessel_params = {
        'isthmus': {
          'isthmus_height': to_meters(np.random.uniform(height/3.75, height/2.4)),
          'isthmus_width': to_meters(np.random.uniform(length/15, length/4)),
          'isthmus_position': to_meters(np.random.uniform(length/4, length - length/4)),
          'undulation_amplitude': to_meters(np.random.uniform(height/40, height/20)),
          'undulation_frequency': np.random.uniform(1, 2),
        }}

    case 'branch':
      vessel_params = {
        'branch': {
          'branch_width': to_meters(np.random.uniform(height, height*1.2)),
          'branch_angle': np.random.uniform(np.pi/8, np.pi/4),
          'branch_position_x': to_meters(length/6),
          'undulation_amplitude': to_meters(np.random.uniform(height/40, height/20)),
          'undulation_frequency': np.random.uniform(1, 3),
          'oscilations': int(np.random.uniform(2, 5)),
      }}

    case 'aneurysm':
      vessel_params = {
        'aneurysm': {
          'aneurysm_width': to_meters(np.random.uniform(height/1.5, height/1.1)),
          'aneurysm_height': to_meters(np.random.uniform(height/3.75, height/1.5)),
          'aneurysm_strecth': to_meters(np.random.uniform(0, height/2.5)),
          'aneurysm_intake': to_meters(np.random.uniform(height/15, height/3.75)),
          'aneurysm_roundness': to_meters(np.random.uniform(height, height/0.75)),
          'aneurysm_center': to_meters(np.random.uniform(height/1.1, length - height/1.1)),
          'undulation_amplitude': to_meters(np.random.uniform(height/40, height/25)),
          'undulation_frequency': np.random.uniform(1, 3),
          'oscilations': int(np.random.uniform(2, 5)),
      }}

  return vessel_params
```

```python
def random_vessel_type():
  vessel_types = ['healthy', 'isthmus', 'branch', 'aneurysm']
  return random.choice(vessel_types)
```
